# The Fundamentals of Machine Learning
## What Is Machine Learning?

## Three ML alternative definitions:
1. It is the science (and art) of programming computers so they can learn from data. (Aurelien Geron, 2018)
2. It is the field of study that gives computers the ability to learn without being explicitly programmed. (Arthur Samuel, 1959)
3. A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E. (Tom Mitchell, 1997)
     * This last definition is more engineering oriented.

## Case 1: a spam filter is a ML program that with examples of spam emails and regular emails can learn to flag spam. 
* The examples that the system uses to learn are called the training set. 
* Each training example is called a training instance (or sample). 
* The task T is to flag spam for new emails, the experience E is the training data, and the performance measure P can be the ratio of correctly classified emails.
    * This particular performance measure is called accuracy, and it is often used in classification tasks.

## Case 2: dowloading a copy of Wikipedia.
* If you download a copy of Wikipedia, your computer has a lot more data, but it is not better at any task.
    * Downloading a copy of Wikipedia is not Machine Learning.

# Why Use Machine Learning?
## Imagine writing a spam filter using traditional programming.

### 1. First you consider spam characteristics:
   ####  - Words or phrases (“4U,” “credit card,” “free,”, “amazing”...) in the subject line.
   ####  - Patterns in the sender name, email body, etc.
### 2. You write a detection algorithm the patterns set 
   ####  - The program will flag emails as spam if a number of these patterns were detected.
### 3. You test the program and repeat steps 1 and 2 until it was good enough.


### Since the problem is difficult, your program will likely become a long list of complex rules—hard to maintain.

# ML Spam-filter program
### A Machine Learning program automatically learns which words and phrases are good predictors of spam.
* It does it detecting frequent patterns of words in the spam examples compared to the good examples.
* The program is shorter, easier to maintain, and most likely more accurate.
    * If spammers notice that all their emails containing “4U” are blocked they might start writing “For U”. A traditional spam filter would need to be updated to flag “For U” emails.
    * A spam filter based on ML automatically notices that “For U” has become unusually frequent in spam flagged by users, and starts flagging them without intervention.

### ML shines in problems that either are too complex for traditional approaches or have no known algorithm. 
## Eg.: speech recognition. A program capable of distinguishing the words “one” and “two”. 
* The word “two” starts with a high-pitch sound (“T”).
* An algorithm that measures high-pitch sound intensity can be coded and used to distinguish ones and twos.
* However, this technique will not scale to thousands of words spoken by millions of people in noisy environments.
## Best solution: write an algorithm that learns by itself, given many example recordings for each word.

## ML can help humans learn. ML algorithms can be inspected to see what they have learned.
### For instance, once a spam filter can be inspected to reveal the list of words and combinations of words that it believes are the best predictors of spam. 
* Sometimes this will reveal unsuspected correlations or new trends and lead to a better understanding of the problem.
* Applying ML to dig into large amounts of data can help discover patterns that were not immediately apparent. This is called data mining.

## In summary, ML is great for:

* Problems for which existing solutions require a lot of fine-tuning or long lists of rules: one ML algorithm can simplify code and perform better than the traditional approach.
* Complex problems for which using a traditional approach yields no good solution: the best ML techniques can perhaps find a solution.
* Fluctuating environments: a ML system can adapt to new data.
* Getting insights about complex problems and large amounts of data.

## Types of Machine Learning Systems
### It is useful to classify ML systems on the following criteria:
* If they are trained or not with human supervision (supervised, unsupervised, semisupervised, and Reinforcement Learning)
* If they can or can not learn incrementally on the fly (online versus batch learning)
* If they work by simply comparing new data points to known data points, or instead by detecting patterns in the training data and building a predictive model (instance- based versus model-based learning)

## Supervised/Unsupervised Learning
## There are four major categories: 
* Supervised learning: the training set you feed to the algorithm includes the desired solutions, called labels.


* A supervised learning task is classification:
    * The spam filter: it is trained with many example emails along with their class (spam or good), and it must learn how to classify new emails.
    * Regression tasks, such as predicting a target numeric value (eg. the price of a car), given a set of features (mileage, age, brand, etc.) called predictors.
        * NOTE: In ML an attribute is a data type (e.g., “mileage”), while a feature has several meanings, depending on the context, but generally means an attribute plus its value (e.g., “mileage = 15,000”).
        * Regression tasks, we shall see, can be used for classification also and the other way around.

## Some supervised learning algorithms:
* k-Nearest Neighbors
* Linear Regression
* Logistic Regression
* Support Vector Machines (SVMs)
* Decision Trees and Random Forests 
* Neural networks

## Unsupervised learning
### In unsupervised learning the training data is unlabeled.
* The system tries to learn without a teacher.


### Some unsupervised learning algorithms:
* Clustering
    * K-Means
    * DBSCAN
    * Hierarchical Cluster Analysis (HCA)
* Anomaly detection and novelty detection
    * One-class SVM

* Visualization and dimensionality reduction

    * Eg., it might find that 40% of your visitors are men who read the blog in the evening, while 20% are young people who visit during the weekends.
* If you use a hierarchical clustering algorithm, it may also subdivide each group into smaller groups. This may help you target your posts for each group.

## Othe unsupervised tasks

    * Simplify the data without losing too much information. 
        * One way is to merge several correlated features into one (feature extraction). 
            * For example, a car’s mileage may be strongly correlated with its age, so the dimensionality reduction algorithm will merge them.

            * It will run much faster, the data will take up less disk and memory space, and in some cases it may also perform better.

* Anomaly detection.

## Semisupervised learning

    * Usually: plenty of unlabeled instances, and few labeled instances.

# Batch and Online Learning

### The second criterion to classify ML systems relies upon whether the system can learn incrementally from a stream of incoming data.

## Batch learning
* In batch learning, the system can not learn incrementally: it must be trained using all the available data.
    * This takes a lot of time and computing resources: so it is done offline.
    * First the system is trained, and then it is launched into production and runs without further learning.
    * If you obtain an improved data set you need to train from scratch.
* Batch learning is optimal for training with big data sets.
    * This is not ideal in small systems (cell phones, a rover on Mars).

## Online learning
* In online learning the system is incrementally trained by feeding it data instances sequentially.
    * Either individually or in small groups called mini-batches.
* Each learning step is fast and cheap.
    * The system can learn about new data on the fly, as it arrives.
    


Online learning is optimal for systems:
* That receive data as a continuous flow and need to adapt to change rapidly or autonomously. 
* With limited computing resources.
    * Once an online learning system has learned about new data instances they can discarded them.
* Trained in huge datasets that cannot fit in one machine’s main memory.
    * This is called out-of- core learning: the algorithm loads part of the data, runs a training step on that data, and repeats the process until it has run on all of the data.
    * WARNING: out-of-core learning is usually done offline and online learning can be a confusing name. Think of it as incremental learning.


* Learning Rate: is the parameter that measures how fast online learning systems adapt to changing data.
* A high learning rate:
    * the system will rapidly adapt to new data,
    * but it will also tend to quickly forget the old data.
* A small learning rate:
    * the system will learn more slowly, 
    * but it will also be less sensitive to noise or outliers.

# A big challenge with online learning
* If bad data is fed to the system, the system’s performance will gradually decline. 
    * For example, bad data could come from a malfunctioning sensor on a robot, or from someone spamming a search engine to try to rank high in search results. 
    * To reduce this risk, you need to monitor your system closely and promptly switch learning off if you detect a drop in performance. 
    * You may also want to monitor the input data and react to abnormal data (e.g., using an anomaly detection algorithm).

# Instance-Based Versus Model-Based Learning
The last way to categorize ML systems is by how they generalize.
* Most ML tasks are about making predictions.
* Given a number of training examples, the system needs to make good predictions for (generalize to) examples it has never seen before. 
* Having a good performance on training data is good, but insufficient; the true goal is to perform well on new instances.

There are two main approaches to generalization:
* Instance-based learning
* Model-based learning

## Instance-based learning
* The most trivial form of learning is to do it by heart.
    * A spam filter "by heart" would flag all emails that are identical to emails that have already been flagged by users.
        * Not the worst solution, but certainly not the best.
* An improvement considers a measure of similarity with monitored instances.
    * In the example, instead of just flagging emails that are identical to known spam emails, the filter could also flag emails that are very similar to known spam emails.
    * A (very basic) similarity measure between two emails could be to count the number of words they have in common.
    * The system would flag an email as spam if it has many words in common with a known spam email.

This is called instance-based learning: the system learns the examples by heart, then generalizes to new cases by using a similarity measure to compare them to the learned examples (or a subset of them).


In the figure the new instance would be classified as a triangle because the majority of the most similar instances belong to that class.


## Model-based learning
Another way to generalize from a set of examples is to build a model of these examples and then use that model to make predictions. This is called model-based learning.


### Example
You want to study the correlation between income and happiness. 
* The OECD provides with a Better Life Index.
* The IMF has the GDP data.
    * A table can be built:



Although there is some noise the data seem to follow a line. A linear **model selection:** with parameters $\theta_0$ and $\theta_1$ can be proposed

\begin{equation}
\text{life_satisfaction} = \theta_0 + \theta_1 \times \text{GDP_per_capita}
\end{equation}

The optimal values for $\theta_0$ and $\theta_1$ are to be found minimizing a cost function that needs to be defined:
* The cost function measures the distance between the linear model predictions and the training examples.




Then a Linear Regression algorithm finds the parameters that make the linear model fit best to data using the training examples.
* This is called training the model.
* In this case an algorithm finds $\theta_0 = 4.85$ and $\theta_1 = 4.91 \times 10^{–5}$.



Model selection consists in choosing the type of model and fully specifying its architecture. Training a model means running an algorithm to find the model parameters that will make it best fit the training data and make good predictions on new data.

This function just merges the OECD's life satisfaction data and the IMF's GDP per capita data.

```python
def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
    oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
    gdp_per_capita.set_index("Country", inplace=True)
    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                                  left_index=True, right_index=True)
    full_country_stats.sort_values(by="GDP per capita", inplace=True)
    remove_indices = [0, 1, 6, 8, 33, 34, 35]
    keep_indices = list(set(range(36)) - set(remove_indices))
    return full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]
```

## Poor-Quality Data
If the training data is full of errors, outliers, and noise it will make it harder for the system to detect the underlying patterns.

* Cleaning up training data is worth the effort.
   * Data scientists spend a significant part of their time doing just that.

Two examples of when want to clean up training data:
* If some instances are clearly outliers, it may help to simply discard them or try to fix the errors manually.
* If some instances are missing a few features (e.g., 5% of your customers did not specify their age), a decision has to be taken about ignoring this attribute, these instances, fill in the missing values (e.g., with the median age), or train one model with the feature and one model without it.


## Irrelevant Features

The system will only be capable of learning if the training data contains enough relevant features
and not too many irrelevant ones. 

Identifying a good data set, called feature engineering, involves three steps:
* Feature selection: selecting the most useful features to train on among existing features.
* Feature extraction: combining existing features to produce a more useful one, here dimensionality reduction algorithms can help.
* Creating new features by gathering new data.

# Overfitting the Training Data

When the model performs well on the training data but it does not generalize well we have an overfitting problem.
* Overfitting happens when the model is too complex relative to the amount and noisiness of the training data.
* Example a very high degree polynomial.


```python
full_country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction', figsize=(8,3))
plt.axis([0, 110000, 0, 10])

from sklearn import preprocessing
from sklearn import pipeline

poly = preprocessing.PolynomialFeatures(degree=60, include_bias=False)
scaler = preprocessing.StandardScaler()
#lin_reg2 = linear_model.LinearRegression()
# Select a linear model
lin_reg2 = sklearn.linear_model.LinearRegression()

pipeline_reg = pipeline.Pipeline([('poly', poly), ('scal', scaler), ('lin', lin_reg2)])
pipeline_reg.fit(Xfull, yfull)
curve = pipeline_reg.predict(X[:, np.newaxis])
plt.plot(X, curve)
plt.xlabel("GDP per capita (USD)")
save_fig('overfitting_model_plot',resolution=100)
plt.show()
```

## Possible solutions to overfitting
* **Regularization**: simplify the model selecting one with fewer parameters, reducing the number of attributes in the training data, or or constraining the model.
* Gather more training data.
* Reduce the noise in the training data (e.g., fix data errors and remove outliers).

## Regularization
The figure shows three models:
* Dotted line: the original model (trained on the countries represented as circles).
* Dashed line: trained with all countries (circles and squares).
* Solid line: trained wiht the circles but with a regularization constraint.
Regularization forced the model to have a smaller slope: it does not fit the training data as well as the first model but generalizes better to new examples.


```python
plt.figure(figsize=(8,3), dpi= 60)

plt.xlabel("GDP per capita")
plt.ylabel('Life satisfaction')

plt.plot(list(sample_data["GDP per capita"]), list(sample_data["Life satisfaction"]), "bo")
plt.plot(list(missing_data["GDP per capita"]), list(missing_data["Life satisfaction"]), "rs")

X = np.linspace(0, 110000, 1000)
plt.plot(X, t0full + t1full * X, "r--", label="Linear model on all data")
plt.plot(X, t0 + t1*X, "b:", label="Linear model on partial data")

ridge = sklearn.linear_model.Ridge(alpha=10**9.5)
Xsample = np.c_[sample_data["GDP per capita"]]
ysample = np.c_[sample_data["Life satisfaction"]]
ridge.fit(Xsample, ysample)
t0ridge, t1ridge = ridge.intercept_[0], ridge.coef_[0][0]
plt.plot(X, t0ridge + t1ridge * X, "b", label="Regularized linear model on partial data")

plt.legend(loc="lower right")
plt.axis([0, 110000, 0, 10])
plt.xlabel("GDP per capita (USD)")
save_fig('ridge_model_plot',resolution=60)
plt.show()
```

The amount of regularization is controlled by a hyperparameter.
* The hyperparameter is a parameter of the learning algorithm, not of the model. 
* It is set prior to training and remains constant.
* A much larger value of the hyperparameter results in a flat model.
    * The learning algorithm will not overfit the training data but it will be unlikely to find a good solution.
* Tuning hyperparameters is an important part of building a ML system.
    * We will go over it in future lectures.

## Underfitting the Training Data
Underfitting is the opposite of overfitting: it arises when the model is too simple to describe the underlying structure of the data.
* For example, a linear model of life satisfaction: reality more complex than the model and its predictions are bound to be inaccurate.

## Possible solutions to underrfitting
* Select a more powerful model, with more parameters.
* Feed better features to the learning algorithm (feature engineering).
* Reduce the constraints on the model (e.g., reduce the regularization hyperparameter).

## Testing and Validating
* To know how a model generalizes it has to be tried on new cases.
    * Applying the model to new data might be risky. 
* A better option is to split the initial data into two sets: the training and test set.
* The error rate on new cases is called the generalization error and tells how well the model will perform on new instances.
    * If the training error is low but the generalization error is high, the model is overfitting the training data.

## Testing and Validating
* Usually 80% of the data are used for training and 20% for testing.
* This depends on the size of the dataset.
* For large samples a smaller testing percentage can be envisaged.

# Hyperparameter Tuning and Model Selection
* Using regularization hyperparameters is a *cure* for overfitting. 
* An optimal hyperparameter value is wanted.
* The choice can be made scanning the hyperparameter space, training different models and then testing in the validating sample.
    * If the validating sample is too small the error rate can be significantly larger when the model is generalized.
    * Is similar to overfitting but affects the validation data and the hyperparameters.

## Holdout validation
* If the validation sample is smaller than the training sample *holdout validation* is used to avoid this effect.
    * Validation of the different parameters is done first in a subset of the trainning sample called the *validation set*.
    * A second performance check is then made in the original *validation sample*.
    * This is the way to go for small validation samples.

## Cross validation
* If the validation sample is large (compared to the training sample) *cross-validation* is applied.
    * The validation sample is split in subsamples of the size of the training sample.
    * Every model is tested against each of the validation subsamples.
    * The average of the error rate is assigned as benchmark index.

## *No Free Lunch*  (NFL) Theorem
A model is a generalization of observations.
* Superfluous details, unlikely to generalize to new instances, are disregarded making assumptions.
    * For example, a linear model assumes that the data are linear and the distance between the instances and the line is noise.

NFL theorem: In a famous 1996 paper, David Wolpert demonstrated that if you
make absolutely no assumption about the data, there is no reason
to prefer one model over any other.
* For some datasets the best model is a Linear Model, while for other datasets it is a Neural Network. 
* There is no model that is a priori guaranteed to work better.

The only way to know for sure which model is best is to evaluate them all.
* Since this is not possible, some reasonable assumptions about the data have to be made and evaluate only a few reasonable models.
* For example, for simple tasks linear models with various levels of regularization.
* For complex problems different kinds of Neural Networks.
