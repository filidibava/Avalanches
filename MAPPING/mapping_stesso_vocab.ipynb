MAPPING CUCITO ADDOSSO

```python
import os
import random
import numpy as np
from collections import Counter
from tqdm import tqdm

def map_sequences(arr):
    """
    Mappa le sequenze UMT con le sequenze originali
    
    Args:
        arr: lista di elementi da usare per completare il mapping
    """
    
    # Cartelle
    original_dir = 'SEQUENCES_LARGER_THAN_500'
    umt_dir = 'UMT_SEQUENCES_LARGER_THAN_500'
    output_dir = 'UMT_SEQUENCES_MAPPED_LARGER_THAN_500_NEW_MAPPING'
    
    # Crea la cartella di output
    os.makedirs(output_dir, exist_ok=True)
    
    # Controlla che le cartelle esistano
    if not os.path.exists(original_dir):
        print(f"Errore: Cartella {original_dir} non trovata")
        return
    
    if not os.path.exists(umt_dir):
        print(f"Errore: Cartella {umt_dir} non trovata")
        return
    
    # Ottieni tutti i file UMT
    umt_files = [f for f in os.listdir(umt_dir) if f.endswith('_UMT.txt')]
    
    processed_count = 0
    error_count = 0
    
    # Progress bar per i file UMT
    for umt_file in tqdm(umt_files[:], desc="Mappando", unit="file"):
        try:
            # Estrai il numero dal nome del file UMT
            # text_numero_UMT.txt -> text_numero.txt
            base_name = umt_file.replace('_UMT.txt', '.txt')
            
            original_path = os.path.join(original_dir, base_name)
            umt_path = os.path.join(umt_dir, umt_file)
            
            # Controlla se il file originale esiste
            if not os.path.exists(original_path):
                error_count += 1
                continue
            
            # Leggi e conta frequenze del file originale
            with open(original_path, 'r', encoding='utf-8') as f:
                original_elements = f.read().split()
            
            original_counter = Counter(original_elements)
            original_sorted = [elem[0] for elem in original_counter.most_common()]
            #print(original_sorted)
            
            # Leggi e conta frequenze del file UMT
            with open(umt_path, 'r', encoding='utf-8') as f:
                umt_elements = f.read().split()
            
            umt_counter = Counter(umt_elements)
            umt_sorted = [elem[0] for elem in umt_counter.most_common()]
            #print(umt_sorted)
            support_dict = {}
            min_len = min(len(original_sorted), len(umt_sorted))

            for i in range(min_len):
                support_dict[umt_sorted[i]] = original_sorted[i]
            
            if len(umt_sorted) > len(original_sorted):
                used_values = set(original_sorted)
                remaining_values = np.setdiff1d(arr, list(used_values), assume_unique=True)
                needed = len(umt_sorted) - len(original_sorted)
                sampled_values = np.random.choice(remaining_values, size=needed, replace=False)

                for j in range(len(original_sorted), len(umt_sorted)):
                    key = umt_sorted[j]
                    support_dict[key] = sampled_values[j - len(original_sorted)]

            mapped_sequence = [str(support_dict.get(token, -1)) for token in umt_elements if token in support_dict]
            
            output_filename = umt_file.replace('_UMT.txt', '_UMT_mapped.txt')
            output_path = os.path.join(output_dir, output_filename)
            
            with open(output_path, 'w', encoding='utf-8') as f:
                f.write('\n'.join(mapped_sequence))
            
            processed_count += 1
            
        except Exception as e:
            error_count += 1
    
    print(f"Completato: {processed_count} successi, {error_count} errori")
```

MAPPING CUCITO ADDOSSO T.C.I VOCABOLARI ABBIANO STESSA CARDINALITÀ TRANNE IL CASO IN CUI D_L > D_SIMULAZIONE

```python
import os
import random
import numpy as np
from collections import Counter
from tqdm import tqdm

def map_sequences(arr):
    """
    Mappa le sequenze UMT con le sequenze originali
    
    Args:
        arr: lista di elementi da usare per completare il mapping
    """
    
    # Cartelle
    original_dir = 'SEQUENCES_LARGER_THAN_500/SEQUENCES'
    umt_dir = 'UMT_SEQUENCES_LARGER_THAN_500'
    output_dir = 'UMT_SEQUENCES_MAPPED_LARGER_THAN_500_NEW_MAPPING'
    
    # Crea la cartella di output
    os.makedirs(output_dir, exist_ok=True)
    
    # Controlla che le cartelle esistano
    if not os.path.exists(original_dir):
        print(f"Errore: Cartella {original_dir} non trovata")
        return
    
    if not os.path.exists(umt_dir):
        print(f"Errore: Cartella {umt_dir} non trovata")
        return
    
    # Ottieni tutti i file UMT
    umt_files = [f for f in os.listdir(umt_dir) if f.endswith('_UMT.txt')]
    
    processed_count = 0
    error_count = 0
    
    # Progress bar per i file UMT
    for umt_file in tqdm(umt_files[:], desc="Mappando", unit="file"):
        try:
            # Estrai il numero dal nome del file UMT
            # text_numero_UMT.txt -> text_numero.txt
            base_name = umt_file.replace('_UMT.txt', '.txt')
            
            original_path = os.path.join(original_dir, base_name)
            umt_path = os.path.join(umt_dir, umt_file)
            
            # Controlla se il file originale esiste
            if not os.path.exists(original_path):
                error_count += 1
                continue
            
            # Leggi e conta frequenze del file originale
            with open(original_path, 'r', encoding='utf-8') as f:
                original_elements = f.read().split()
            
            original_counter = Counter(original_elements)
            original_sorted = [elem[0] for elem in original_counter.most_common()]
            D_l = len(original_sorted)  # Numero di elementi distinti nella sequenza originale
            
            # Leggi il file UMT
            with open(umt_path, 'r', encoding='utf-8') as f:
                umt_elements = f.read().split()
            
            # Calcola D_s totale
            umt_counter = Counter(umt_elements)
            D_s = len(umt_counter)  # Numero di elementi distinti nella simulazione
            
            # Gestisci i tre casi
            if D_s > D_l:
                # Caso D_s > D_l: tronca la simulazione
                truncated_elements = []
                seen_elements = set()
                
                for element in umt_elements:
                    if element not in seen_elements:
                        # Nuovo elemento distinto
                        if len(seen_elements) == D_l:
                            # Abbiamo già D_l elementi distinti, fermiamoci
                            break
                        seen_elements.add(element)
                    truncated_elements.append(element)
                
                # Usa gli elementi troncati per il mapping
                umt_elements_to_use = truncated_elements
                umt_counter = Counter(umt_elements_to_use)
                
            else:
                # Caso D_s <= D_l: usa tutta la sequenza
                umt_elements_to_use = umt_elements
            
            # Procedi con il mapping normale
            umt_sorted = [elem[0] for elem in umt_counter.most_common()]
            
            support_dict = {}
            min_len = min(len(original_sorted), len(umt_sorted))

            for i in range(min_len):
                support_dict[umt_sorted[i]] = original_sorted[i]
            
            if len(umt_sorted) > len(original_sorted):
                used_values = set(original_sorted)
                remaining_values = np.setdiff1d(arr, list(used_values), assume_unique=True)
                needed = len(umt_sorted) - len(original_sorted)
                sampled_values = np.random.choice(remaining_values, size=needed, replace=False)

                for j in range(len(original_sorted), len(umt_sorted)):
                    key = umt_sorted[j]
                    support_dict[key] = sampled_values[j - len(original_sorted)]

            mapped_sequence = [str(support_dict.get(token, -1)) for token in umt_elements_to_use if token in support_dict]
            
            output_filename = umt_file.replace('_UMT.txt', '_UMT_mapped.txt')
            output_path = os.path.join(output_dir, output_filename)
            
            with open(output_path, 'w', encoding='utf-8') as f:
                f.write('\n'.join(mapped_sequence))
            
            processed_count += 1
            
        except Exception as e:
            error_count += 1
    
    print(f"Completato: {processed_count} successi, {error_count} errori")
```
