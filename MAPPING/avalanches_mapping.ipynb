```python
def delays(novelties,flag):
    delay = np.diff(novelties)
    delay = np.array([i for i in delay if i!=1])
    delay = np.array([i-1 for i in delay])
    if(flag == True):
        return np.log(delay)
    else:
        return delay
```

```python

import os
import re
from tqdm import tqdm

def extract_number_from_filename(filename):
    """Estrae il numero dal nome del file per l'ordinamento"""
    match = re.search(r'text_(\d+)', filename)
    return int(match.group(1)) if match else 0

def stack_files_in_order(mapped_dir, original_dir, 
                        mapped_output="UMT_mapped_stacked_from_zipf.txt", 
                        wiki_output="wiki_stacked.txt"):
    """
    Unisce i file in ordine numerico mantenendo la corrispondenza
    
    Args:
        mapped_dir: cartella con file text_numero_UMT_mapped.txt
        original_dir: cartella con file text_numero.txt
        mapped_output: nome file output per i mapped
        wiki_output: nome file output per i wiki
    """
    
    # Controlla che le cartelle esistano
    if not os.path.exists(mapped_dir):
        print(f"Errore: Cartella {mapped_dir} non trovata")
        return
    
    if not os.path.exists(original_dir):
        print(f"Errore: Cartella {original_dir} non trovata")
        return
    
    # Trova tutti i file mapped
    mapped_files = [f for f in os.listdir(mapped_dir) if f.endswith('_UMT.txt')]
    
    # Ordina i file per numero
    mapped_files.sort(key=extract_number_from_filename)
    
    print(f"Trovati {len(mapped_files)} file da unire in ordine...")
    
    # Crea le coppie corrispondenti
    file_pairs = []
    missing_files = []
    
    for mapped_file in mapped_files:
        # Estrai il nome del file originale corrispondente
        original_file = mapped_file.replace('_UMT.txt', '.txt')
        original_path = os.path.join(original_dir, original_file)
        
        if os.path.exists(original_path):
            file_pairs.append((mapped_file, original_file))
        else:
            missing_files.append(original_file)
    
    if missing_files:
        print(f"Attenzione: {len(missing_files)} file originali mancanti:")
        for missing in missing_files[:5]:  # Mostra solo i primi 5
            print(f"  - {missing}")
        if len(missing_files) > 5:
            print(f"  ... e altri {len(missing_files) - 5}")
    
    print(f"Unendo {len(file_pairs)} coppie di file in ordine numerico...")
    
    # Unisci i file mapped
    with open(mapped_output, 'w', encoding='utf-8') as mapped_out:
        for mapped_file, _ in tqdm(file_pairs, desc="Unendo file mapped"):
            mapped_path = os.path.join(mapped_dir, mapped_file)
            
            try:
                with open(mapped_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                    # Rimuovi righe vuote e aggiungi contenuto
                    lines = [line for line in content.split('\n') if line.strip()]
                    mapped_out.write('\n'.join(lines) + '\n')
                    
            except Exception as e:
                print(f"Errore leggendo {mapped_file}: {e}")
    
    # Unisci i file originali (wiki)
    with open(wiki_output, 'w', encoding='utf-8') as wiki_out:
        for _, original_file in tqdm(file_pairs, desc="Unendo file wiki"):
            original_path = os.path.join(original_dir, original_file)
            
            try:
                with open(original_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                    # Rimuovi righe vuote e aggiungi contenuto
                    lines = [line for line in content.split('\n') if line.strip()]
                    wiki_out.write('\n'.join(lines) + '\n')
                    
            except Exception as e:
                print(f"Errore leggendo {original_file}: {e}")
    
    # Verifica le lunghezze finali
    try:
        with open(mapped_output, 'r') as f:
            mapped_lines = len(f.readlines())
        
        with open(wiki_output, 'r') as f:
            wiki_lines = len(f.readlines())
        
        print(f"\nRisultati:")
        print(f"✓ {mapped_output}: {mapped_lines:,} righe")
        print(f"✓ {wiki_output}: {wiki_lines:,} righe")
        
        if mapped_lines == wiki_lines:
            print("✓ Le lunghezze corrispondono!")
        else:
            print(f"⚠️  Attenzione: lunghezze diverse ({mapped_lines - wiki_lines:+,})")
            
    except Exception as e:
        print(f"Errore verificando le lunghezze: {e}")
    
    # Mostra l'ordine dei primi file
    print(f"\nOrdine di unione (primi 10 file):")
    for i, (mapped_file, original_file) in enumerate(file_pairs[:10], 1):
        number = extract_number_from_filename(mapped_file)
        print(f"  {i:2d}. text_{number}.txt + text_{number}_UMT_mapped.txt")
    
    if len(file_pairs) > 10:
        print(f"  ... e altri {len(file_pairs) - 10} file")
    
    print(f"\nFile creati:")
    print(f"  - {mapped_output}")
    print(f"  - {wiki_output}")

if __name__ == "__main__":
    # Configura i path delle cartelle
    MAPPED_DIR = "SEQUENZE_MAPPATE"
    ORIGINAL_DIR = "SEQUENCES_LARGER_THAN_500/SEQUENCES"
    
    stack_files_in_order(MAPPED_DIR, ORIGINAL_DIR)
```
